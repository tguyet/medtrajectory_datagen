## Identification des pharmacies

```python
#catégories affinées des établissements "Commerce de Biens à Usage Médicaux"
data[data['categagretab']==3201][["categetab", "libcategetab"]].drop_duplicates()
```

```python
data["departement"].unique()
```

```python
#selection des pharmacies du département 35
data35 = data[ (data["departement"]=="35") & (data['categagretab']==3201)]
len(data35)
```

```python
data.columns
```

```python
data35[['nofinesset','rs', 'numvoie', 'typvoie', 'voie', 'commune', 'departement','categetab', 'ligneacheminement']]
```

```python
data35_sc=pd.concat( (data35[['nofinesset','rs', 'numvoie', 'typvoie', 'voie', 'commune', 'departement','categetab']],data35['ligneacheminement'].str.split(n=1,expand=True)), axis=1 )
data35_sc=data35_sc.rename(columns={0:"CP",1:"libcom"})
data35_sc
```

```python
dpts=[22, 35,49]
data["numvoie"].fillna(0, inplace=True)
data["numvoie"]=data["numvoie"].astype(int)
data["commune"].fillna(0, inplace=True)
data["commune"]=data["commune"].astype(int)
data["categetab"].fillna(0, inplace=True)
data["categetab"]=data["categetab"].astype(int)
ddpt=data[ (data["departement"].astype("str")==str(dpts[0])) & (data['categagretab']==3201)]
for dpt in dpts[1:]:
    ddpt = pd.concat( (data[ (data["departement"].astype("str")==str(dpt)) & (data['categagretab']==3201)],ddpt) )
ddpt=pd.concat( (ddpt[['nofinesset','rs', 'numvoie', 'typvoie', 'voie', 'commune', 'departement','categetab']],ddpt['ligneacheminement'].str.replace(pat=" CEDEX", repl="").str.split(n=1,expand=True)), axis=1 )
ddpt=ddpt.rename(columns={0:"CP",1:"libcom"})
```

## Identification des centres hospitaliers

```python
etabldpt=data[ ((data['categagretab']==1101) | (data['categagretab']==1102) | (data['categagretab']==1106) | (data['categagretab']==1110) ) & (data["departement"].astype("str")==str(35)) ]
etabldpt = pd.concat( (etabldpt[['nofinesset', 'rs', 'numvoie', 'typvoie', 'voie', 'commune', 'departement', 'codemft', 'libmft', 'codesph', 'libsph']], etabldpt['ligneacheminement'].str.replace(pat=" CEDEX", repl="").str.split(n=1,expand=True)), axis=1 )
etabldpt.rename(columns={0:"CP",1:"libcom"}, inplace=True)
etabldpt.head()
```
